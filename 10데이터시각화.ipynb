```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 여러 개의 그래프 그리기
+ subplot(행, 열, 번호)


### 좌우로 나란히 그래프 그리기

```python
x = [1,2,3,4]
y = [2,3,5,10]

plt.subplot(1,2,1) # 1행 2열 영역 중 1행 1열 영역
plt.plot(x,y,'r--')

plt.subplot(1,2,2) # 1행 2열 영역 중 1행 2열 영역
plt.plot(x,y,'b')

plt.tight_layout()  # 보기좋게 그래프 크기 재조정
```

### 위아래로 나란히 그래프 그리기

```python
x = [1,2,3,4]
y = [2,3,5,10]

plt.subplot(2,1,1) # 1행 2열 영역 중 1행 1열 영역
plt.plot(x,y,'r--')

plt.subplot(2,1,2) # 1행 2열 영역 중 1행 2열 영역
plt.plot(x,y,'b')

plt.tight_layout()  # 보기좋게 그래프 크기 재조정
```

## 그래프에 텍스트 삽입하기
+ text(x,y, 텍스트)
+ annotate(텍스트,(x,y), 스타일)

```python
x = [1,2,3,4]
y = [2,3,5,10]

plt.scatter(x,y)

style = {'color':'red','weight':'bold'}
plt.text(1,2, 2, fontdict=style)
plt.text(2,3, 3, fontdict=style)
plt.text(3,5, 5, fontdict=style)
plt.text(4,10, 10, fontdict=style)


plt.tight_layout()
```

```python
x = [1,2,3,4]
y = [2,3,5,10]

plt.scatter(x,y)

style = {'color':'red','weight':'bold'}
for i in range(len(x)):
    if i == (len(x)-1): 
        plt.text(x[i]-0.2,y[i],y[i], fontdict=style)
    else: 
        plt.text(x[i],y[i]+0.2,y[i], fontdict=style)

plt.tight_layout()
```

```python
colors = ['red','orange','yellow','green']
plt.bar(x,y,color=colors)

style = {'color':'red','weight':'bold'}
for i in range(len(x)):
     plt.text(x[i]-0.05,y[i]+0.1,y[i], fontdict=style)

plt.show()
```

## 타이타닉 성별 생존자 시각화

```python
titanic = sns.load_dataset('titanic')
```

```python
ax = sns.countplot(data=titanic, x='alive',hue='sex')

for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x()+0.15,p.get_height()+5 ),
               color='red',weight='bold')

plt.show()
```
