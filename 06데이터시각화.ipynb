```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 선그래프
* 시간의 따른 데이터의 변화를 시각화하는데 유용하게 사용
  + 지난 10년 간 경유 가격의 평균값
  + 지난 두 달간 몸무게 변화
* 시계열 데이터를 시각화하는데 주로 사용
* plot(x, y, 옵션)


### 항공기 이용승객 데이터

```python
air = pd.read_csv('data/airpassengers.csv')
air.head()
```

```python
plt.figure(figsize=(25,8))  # 그래프 영역지정 (단위: 인치)
plt.plot(air.Month,air.Passengers)
plt.xticks(rotation=90, fontsize=15)
plt.show()
```

## seaborn에서 선그래프 그리기
+ lineplot(x,y,data)

```python
plt.figure(figsize=(25,8))
sns.lineplot(data=air,x='Month',y='Passengers')
plt.xticks(rotation='vertical')
plt.show()
```

### 영국왕들의 수명데이터

```python
kings = pd.read_csv('data/kings.txt', sep = "\t", header=None)
kings.columns=['age']
```

```python
plt.plot(range(1,len(kings.age)+1),kings.age)
plt.xticks(np.arange(1,len(kings.age)), rotation=90)
plt.show()
```
