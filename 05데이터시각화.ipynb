```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 파이 차트
* 하나의 원을 여러 영역 또는 조각으로 나눈 원그래프
* 각 조각은 해당 변수에 대한 관측치 개수 또는 백분율 의미
* 전체 대비 여러 구성 요소 간 관계를 표시하는 데 유용
* 고객 유형별 백분율, 다양한 제품별 매출 백분율, 다양한 국가별 수익
+ pie(값,레이블,옵션)

```python
# autopct : 부채꼴 안에 백분율 표시
vals = [10,20,30,40]
labels = ['Java','Python','R','C#']

plt.pie(vals, labels=labels, autopct='%d%%')
plt.show()
```

```python
# startangle : 부채꼴이 그려질 시작각도 지정
# 양수 : 시계 반대 방향
# 음수 : 시계 방향
plt.pie(vals, labels=labels, autopct='%d%%', startangle=55)
plt.show()
```

```python
# explode : 부채꼴들이 그래프 중심에서 멀어지게 함
# shadow : 부채꼴 아래 그림자 표시
explodes = [.1,.1,.1,.1]
plt.pie(vals, labels=labels, autopct='%d%%',
       explode=explodes, shadow=True)
plt.show()
```

```python
piestyle = {'width':.7, 'edgecolor':'k','linewidth':.5}
colors = ['red','orange','yellow','green']
explodes = [.1,.1,.1,.1]
plt.pie(vals, labels=labels, autopct='%d%%',
       explode=explodes, colors = colors,
       wedgeprops=piestyle)
plt.show()
```

## 사원들의 직책별 분포시각화

### 방법 1

```python
emp = pd.read_csv('data/EMPLOYEES.csv')
```

```python
vals = emp.JOB_ID.value_counts().values

labels = emp.JOB_ID.value_counts().index
```

```python
plt.pie(vals[:8], labels=labels[:8], autopct='%.1f%%')
plt.show()
```

### 방법 2

```python
jobid = emp.JOB_ID

df = pd.DataFrame(jobid.value_counts())
```

```python
df['prop'] = df.JOB_ID/np.sum(df.JOB_ID)
```

```python
colors = sns.color_palette('Paired')
sns.countplot(x=jobid,  palette=colors)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.pie(df.JOB_ID[:8], labels=df.JOB_ID.index[:8], autopct='%.1f%%', counterclock=False, startangle=90)
plt.show()
```

## 타이타닉 승객의 성별 시각화

### 방법 1

```python
titanic = pd.read_csv('data/titanic.csv')
```

```python
vals = titanic.sex.value_counts().values

labels = titanic.sex.value_counts().index
```

```python
plt.pie(vals, labels=labels, autopct='%d%%')
plt.show()
```

### 방법 2

```python
gender = titanic.sex
df = pd.DataFrame(gender.value_counts())
```

```python
# 항목별 백분율 계산
df['prop'] = df.sex/np.sum(df.sex)
df
```

```python
sns.countplot(data=titanic, x='sex')
plt.show()
```

```python
# counterclock : 그래프를 시계 반대방향으로 그리는 여부 지정
plt.pie(df.prop, labels=df.sex.index, autopct='%.1f%%',startangle=90,
        counterclock=False, shadow=True)
plt.show()
```
