## 명목형 데이터 시각화
+ 빈도에 따른 막대그래프
+ 빈도비율에 따른 원그래프

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 사원들의 연봉 직책 시각화

```python
emp = pd.read_csv('data/EMPLOYEES.csv')
```

```python
# 사원들의 직책에 대한 빈도 계산
# emp.JOB_ID.value_counts()  # 결과는 Series 객체로 출력

df = pd.DataFrame(emp.JOB_ID.value_counts())
df
```

```python
# 빈도를 막대그래프로 시각화
# bar(x축, 크기, 옵션)
colors = ['red','orange','yellow','green','blue','navy','purple']
plt.bar(df.JOB_ID.index, df.JOB_ID, color=colors)
plt.xticks(rotation='vertical')
plt.show()
```

## Pandas에서 바로 시각화하기
+ 객체명.plot.bar

```python
df.plot.bar(color=colors)
plt.show()
```

## seaborn으로 빈도 시각화
+ countplot
+ https://matplotlib.org/stable/tutorials/colors/colormaps.html >> 색깔 규정

```python
# seaborn네서 제공하는 팔레트는 총 6가지
# deep, bright, dark, muted, pastel, colorblind
# palette = sns.color_palette('bright')   # 내장
palette = sns.color_palette('Paired',9)   # 커스텀
```

```python
sns.barplot(data = df,x=df.JOB_ID.index,y= df.JOB_ID,palette=palette)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.countplot(data=emp,x='JOB_ID', palette=palette)
plt.xticks(rotation='vertical')
plt.show()
```

### 교통사고 현황 데이터 시각화

## Pandas에서 특정 컬럼 추출
+ pandas 객체에서 특정컬럼들을 선택할때는 리스트 slicing 또는 iloc/loc 이용
+ 객체명.iloc[:,[컬럼인덱스,...]]
+ 객체명.iloc[,[컬럼],...]

```python
car = car.loc[:,['발생지시도','요일']]
car.head()
```

```python
car = car.iloc[:,[10,4]]
car.head()
```

```python
# 발생지시도별 교통사고 발생건수
```

```python
# 1
df = pd.DataFrame(car.발생지시도.value_counts())
plt.bar(df.발생지시도.index,df.발생지시도,color=colors)
plt.xticks(rotation='vertical')
plt.title('발생지시도별 교통사고 건수')
plt.show()
```

```python
# 2
sns.barplot(data = car,x=car.발생지시도,y=car.발생지시도.index,palette=palette)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# 3
sns.countplot(data=car,x='발생지시도', palette=palette)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# 요일별 교통사고 발생건수
```

```python
# 1
df = pd.DataFrame(car.요일.value_counts())
plt.bar(df.요일.index,df.요일,color=colors)
plt.xticks(rotation='vertical')
plt.title('요일별 교통사고 건수')
plt.show()
```

```python
# 2
sns.barplot(data = car,x=car.요일,y=car.요일.index,palette=palette)
plt.show()
```

```python
# 3
sns.countplot(data=car,x='요일', palette=palette)
plt.xticks(rotation='vertical')
plt.show()
```
